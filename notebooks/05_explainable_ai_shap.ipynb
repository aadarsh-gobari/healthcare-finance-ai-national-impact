```python
# importing necessary libraries
import pandas as pd
import numpy as np
import shap
import joblib
import matplotlib.pyplot as plt
```

```python
# Loading Clean Data and Model
X = pd.read_csv('../data/X_clean.csv')
rf_model = joblib.load('../models/random_forest_healthcare.pkl')
```

```python
# Computing Shap Values
# Use a representative sample for SHAP
X_sample = X.sample(n=200, random_state=42)
```

```python
# Creating an Explainable Model
explainer = shap.Explainer(
    rf_model.predict_proba,
    X_sample,
    algorithm="permutation"
)
```

```python
# Compute Shap values
shap_values = explainer(X_sample)
```

```python
# GGlobal Feature Importance
shap.plots.bar(shap_values[:, :, 1])


```

```python
# Detailed Visualization
shap.plots.beeswarm(shap_values[:, :, 1])
```

### Explainable AI Insights

Explainable AI analysis using permutation-based SHAP identified key clinical features such as age, cholesterol levels, and heart rate as major contributors to disease risk prediction.

This improves transparency, supports clinician trust, and ensures ethical AI adoption in healthcare decision-making.
